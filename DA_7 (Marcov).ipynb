7) Визуализировать  графовые моделеи (networkX) для данного блокнота

```python
import networkx
from networkx import *

gr = Graph()
gr
```

```python
gr.add_node('a')
gr.add_node('b')
gr.add_node('c')

gr.nodes()
```

```python
def add_edge(f_item, s_item, graph=None):
  gr.add_edge(f_item, s_item)
  gr.add_edge(s_item, f_item)
```

```python
 add_edge('a', 'b', graph=graph)
add_edge('b', 'c', graph=graph)
add_edge('b', 'd', graph=graph)
add_edge('b', 'e', graph=graph)
add_edge('a', 'a', graph=graph)

nx.draw_circular(gr,
         node_color='green',
         node_size=1000,
         with_labels=True)   
```
